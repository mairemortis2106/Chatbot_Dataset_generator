```python
from pypdf import PdfReader

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    full_text = ""
    for page in reader.pages:
        full_text += page.extract_text() + "\n" # Extract text from each page
    return full_text

# Example usage:
pdf_file_path = "D:\\GitHub\\Chatbot_Dataset_generator\\pdf_tester\\Kelas_11_SMA_Sejarah_Indonesia_Siswa.pdf" # Replace with your PDF file's path
text = extract_text_from_pdf(pdf_file_path)
print(text)
```

```python
from chunker import chunk_text
from embedding import embed_chunks

chunks = chunk_text(text)
vectors = embed_chunks(chunks)
```

```python
import uuid
from qdrant_client.models import PointStruct
points = [
        PointStruct(id=str(uuid.uuid4()), vector=vector, payload={"text": chunk})
        for chunk, vector in zip(chunks, vectors)
        ]


print(points[:5])
print((len(points)))
```
